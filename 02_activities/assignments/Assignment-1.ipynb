```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y= np.random.randint(0,100,50)


fig, ax = plt.subplots(figsize=(5, 3))
#ax.scatter(x,y)
#ax.bar(x,y)
#ax.barh(x,y)
#ax.plot(x,y)
ax.hist(y)
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20} # Defined a dictionary for the title font: sans-serif family, blue color, and size 20
font2 = {'family':'monospace','color':'green','size':14} # Defined a dictionary for the axis labels: monospace font, green color, and size 14

fig, ax = plt.subplots(figsize=(5, 3)) # Created a figure and axis object with a specific size (5 inches by 3 inches)
ax.plot(x,y)

ax.set_title('Total growth over time',fontdict = font1, loc ='left')
ax.set_ylabel('Total growth',fontdict = font2)
ax.set_xlabel('Years since start',fontdict = font2)
```
