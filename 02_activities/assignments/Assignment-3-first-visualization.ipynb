```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/aktur/Downloads/ontario_top_baby_names_male.csv")
df.columns = ['Year', 'Name', 'Count'] 

latest_year = df['Year'].max()
top5_2023 = df[df['Year'] == latest_year].nlargest(5, 'Count')

name_to_plot = top5_2023.iloc[0]['Name']
name_trend = df[df['Name'] == name_to_plot]

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
fig.suptitle("Ontario Male Baby Names", fontsize=16)
sns.barplot(data=top5_2023, x='Name', y='Count', ax=ax1, palette='Blues_d')
ax1.set_title(f"Top 5 Names in {latest_year}")
ax1.set_ylabel("Number of Babies")
ax1.set_xlabel("Name")
sns.lineplot(data=name_trend, x='Year', y='Count', ax=ax2, marker='o', color='orange')
ax2.set_title(f"Trend of '{name_to_plot}' Over Time")
ax2.set_ylabel("Number of Babies")
ax2.set_xlabel("Year")
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
