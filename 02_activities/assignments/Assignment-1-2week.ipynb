```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
                            nrows=1,
                            figsize=(9, 3))

font1 = {'family' : 'serif','size':10}
font2= {'family' : 'serif', 'size':14}


ax1.scatter(x1,y1,
            
            color= 'red')
ax2.bar(x2,y2, color = 'darkgreen')

ax1.set_title('Scatterplot', fontdict = font2)
ax1.set_ylabel('Measure', fontdict= font1)
ax1.set_xlabel('Time', fontdict = font1)
ax2.set_title('Pirate Strenght', fontdict = font2)
ax2.set_ylabel('Stenght', fontdict = font1)
ax2.set_xlabel('Names', fontdict= font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid')

fig = plt.subplots(figsize=(10, 3))
tipgraph =sns.lineplot(data=tips,
        x='total_bill',
        y='tip',
        color = 'hotpink',
        linestyle = '--',
        linewidth = 3,
        marker = 'o',
        markerfacecolor = 'indigo')

tipgraph.set(title='Tips vs. Total Bill',
            xlabel='Total Bill ($)',
            ylabel='Tip Amount ($)')
```

```python
tipgraph = sns.scatterplot(data=tips, x='total_bill',
                            y='tip', style = 'time', hue =
                            'day', palette = ['purple','hotpink', 'deepskyblue','yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
            xlabel='Total Bill ($)',
            ylabel='Tip Amount ($)')

```

```python
sns.pairplot(data = tips,
            hue = 'day')
```

```python
daysplot = sns.relplot(
            data=tips,
            x="total_bill",
            y="tip",
            hue="sex",
            col="day",
            kind="scatter",
            col_wrap=2)

daysplot
```

```python
import plotly.graph_objects as go

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python

from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                on_bad_lines='skip')
df
42
```

```python
text = " ".join(each for each in df.quote)

wordcloud = WordCloud(background_color="white",
                    colormap = 'inferno').generate(text)

fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud,interpolation='bilinear')
ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B],
                            set_labels = ('Fruits', 'Pies'),
                            set_colors=("blue", "red"),
                            alpha=0.5)

diagram.get_label_by_id("10").set_text("n".join(A - B))

diagram.get_label_by_id("11").set_text("n".join(A & B))

diagram.get_label_by_id("01").set_text("n".join(B - A))
```
